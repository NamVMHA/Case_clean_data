# Tiêu chí 3: Trường đại học

Trước hết ta đọc Dictionary for University_info để xác định các trường có chương trình học "Computer And Information Sciences And Support Services" và ở khu vực thành thị. Ta giữ lại các cột liên qua và bỏ đi các cột thừa
## 1. Các trường ở thành thị
- Đầu tiên ta lọc ra các trường ở 3 khu vực là **San Jose, Los Angeles và San Diego**
- Trong bảng **University_info** ta có cột LOCALE. Trong cột này ta giữ lại các trường đại học có giá trị là **11** (City: Large - population of 250,000 or more) đây là các trường đại học nằm ở các thành phố lớn.

## 2. Chương trình học
- Ta đã xác định được các trường có chương trình học **"Computer And Information Sciences And Support Services"** thường sẽ có dữ liệu trong cột **PCIP11** (Percentage of degrees awarded in Computer And Information Sciences And Support Services) của bảng dữ liệu University_info. Đây là các trường có đào tạo và phát bằng cấp cho Ngành học - Computer And Information Sciences And Support Services mà Robert muốn theo học.
- Tuy nhiên có một số trường đại học cột **PCIP11 giá trị bằng 0** do đó ta sẽ dựa vào **5 cột (CIP11CERT1, CIP11CERT2, CIP11ASSOC, CIP11CERT4, CIP11BACHL) để chọn thêm trường cho Robert**. Đây là các cột thể hiện các bằng cấp chứng chỉ mà trường đó đào tạo với các giá trị 0, 1 ,2 (0: Ko có chương trình, 1: Có chương trình học, 2: Đào tạo từ xa).
- Ta giữ các cột có liên quan đến tiêu chí chọn trường của Robert và loại bỏ bớt các cột không cần thiết 
- Ta xóa các cột không cần thiết trong bảng dữ liệu và chỉ để lại các dữ liệu có liên quan đến ngành học để import vào Python

```python
#import dữ liệu các trường đại học
df_university = pd.read_csv(r'D:\Data Analyst\Learn\MindX\BI12\Final_test\1.Data Wrangling\data\University_info.csv') # / kí tự đặt biệt raw/ rough (để r ở đầu) và file.csv ở cuối
```

```python
# 5 dòng đầu của dữ liệu
df_university.head()
```

```python
# Lấy ra những cột cần thiết cho phân tích
df_university_filter = df_university.filter(["UNITID", "INSTNM", "CITY",'st_fips','region','LOCALE','PCIP11','CIP11CERT1','CIP11CERT2','CIP11ASSOC','CIP11CERT4','CIP11BACHL'])
df_university_filter.head()
```

```python
# lọc ra các trường đại học tương ứng với 3 thành phố San Jose, Los Angeles và San Diego trong tiêu chí
df_uni = df_university_filter.loc[df_university_filter['CITY'].isin(['San Jose','San Diego','Los Angeles'])]  
# lọc ra các trường đại học ở thành thị với giá trị LOCALE = 11
df_uni2 = df_uni.loc[df_uni['LOCALE'] == 11 ] 
# Lọc ra các trường có đào tạo ngành học của Robert (các cột có giá trị > 0: có chương trình thuộc ngành Computer And Information Sciences And Support Services ) 
df_uni3 = df_uni2.loc[(df_uni2['CIP11CERT1'] > 0) | (df_uni2['CIP11CERT2'] > 0) |(df_uni2['CIP11ASSOC'] > 0) |(df_uni2['CIP11CERT4'] > 0) |(df_uni2['CIP11BACHL'] > 0)]
df_uni3
```

```python
#Thông tin dữ liệu
df_uni3.info()
```

```python
#Xem dữ liệu có bị null không
df_uni3.isnull().sum()
```

Như vậy có tổng cộng **28 trường đại học trong 3 thành phố là San Jose, Los Angeles và San Diego (Tất cả đều ở bang California)** đạt tiêu chí của Robert

## Tạo bảng danh sách các trường đại học phù hợp với tiêu chí của Robert

```python
# Danh sách các trường đại học phù hợp với tiêu chí của Robert
df = df_uni3.filter(["CITY", "INSTNM", "UNITID"])
df
```

```python
# Đổi tên các cột trong các bảng để join dữ liệu
df_crime_rename = df_crime_fil3.rename(columns ={'City':'CITY'})
# Lấy ra những cột cần thiết
crimee = df_crime_rename.filter(["CITY", "TotalofCrime"])
crime = crimee.loc[(crimee['TotalofCrime'] < 5706.9)]
crime #Tổng số vụ tội phạm trong các thành phố mà Robert sẽ theo học
```

```python
# Đổi tên cột trong bảng xếp hạng Startup Rank của các thành phố
df_city_fil2 = df_city_fil.rename(columns ={'Metro Area Main City':'CITY'})

# lọc ra thành phố phù hợp với tiêu chí
city_filter = df_city_fil2.filter(["CITY", "Metro Area States",'Startup Rank'])
city = city_filter.loc[(city_filter['CITY'] == 'LosAngeles') | (city_filter['CITY'] == 'SanDiego') | (city_filter['CITY'] == 'SanJose')]                      

# Đổi lại định dạng cho thành phố
city["CITY"].replace({"LosAngeles": "Los Angeles", "SanDiego": "San Diego","SanJose": "San Jose" }, inplace=True)
city
```

```python
# combine data
robert = pd.merge(df,crime,on='CITY', validate='many_to_many')
robert_university = pd.merge(robert,city,on='CITY', validate='many_to_many')
# Bảng tổng hợp các trường đại học theo tiêu chí của Robert
robert_university
```
