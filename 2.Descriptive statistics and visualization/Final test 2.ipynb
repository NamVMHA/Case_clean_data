<h1><font color = 'orange'> Histogram <font></h1>

```python
# Biến động giá nhà tại mỹ
sns.histplot(df['Home Value']) # distplot, histplot
sns.set_style('darkgrid')
plt.title('Home Value frequency histogram', fontsize=14)
plt.ylabel('Frequency', fontsize=10)
plt.xlabel('Home Value', fontsize=10)
```

**Giá nhà tại Mỹ giao động trong khoảng từ 94500 USD đến 537400 USD.**

```python
# Tình hình thu nhập hộ gia đình tại Mĩ
a = df[['HH Inc']]
plt.hist(a)
plt.title('HH Inc frequency histogram', fontsize=14)
plt.ylabel('Frequency', fontsize=10)
plt.xlabel('HH Inc', fontsize=10)
```

**Tổng thu nhập trung bình của các hộ gia đình tại Mỹ giao động trong khoảng từ 37881 USD/năm đến 70647 USD/năm.**

```python
#  thu nhập trên đầu người mỗi gia đình ở Mỹ
b = df[['Per Cap Inc']]
n, bins, patches = plt.hist(b, bins='auto', color='#4876c0',
                            alpha=0.7, rwidth=0.85)
plt.grid(axis='y', alpha=0.75)
plt.xlabel('Per Cap Inc')
plt.ylabel('Frequency')
plt.title('Frequency Histogram Per Cap Inc')
```

**Tổng thu nhập Tổng thu nhập bình quân đầu người tại Mỹ giao động trong khoảng từ 19977 USD/năm đến 42078 USD/năm.**

```python
# tỷ lệ phần trăm số nhà có chủ sở hữu - Pct Owner Occ tại Mỹ
df[['Pct Owner Occ']].plot.hist(grid=True, bins=20, rwidth=0.9,
                   color='#607c8e')
plt.title('Pct Owner Occ - Frequency Histogram')
plt.xlabel('Pct Owner Occ')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
```

**Tỷ lệ phần trăm số nhà có chủ sở hữu tại Mỹ là giao động trong khoảng từ 43.5 % đến 74.6 %.**
